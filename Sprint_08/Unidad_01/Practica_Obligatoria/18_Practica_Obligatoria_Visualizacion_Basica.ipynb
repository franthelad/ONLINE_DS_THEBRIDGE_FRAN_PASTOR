Ax1 Arr-Izq

```python
# voy a crear un df con las 5 ciudades mas pobladas para trabajar con el unicamente y no con las 400 y pico ciudades del df limpio

df_ciudades = df_limpio[df_limpio['city'].isin(['LosAngeles', 'SanDiego', 'Sacramento', 'Anaheim', 'SantaAna'])]

# Ordenamos por poblacion

df_ciudades.sort_values(by= 'population_total', ascending= False, inplace= True)

# ajusto los nombres a la grafica de muestra

cities = ['Los Angeles', 'San Diego', 'Sacramento Ciudades', 'Anaheim', 'Santa Ana']

for index, city in enumerate(df_ciudades['city']):
    df_ciudades.loc[df_ciudades['city'] == city, 'city'] = cities[index]

df_ciudades
```

```python
formatter = ticker.ScalarFormatter(useOffset=False, useMathText=True)
formatter.set_scientific(False)  # desactivar la notación científica

ax1 = ax[0, 0]
bars = (ax1.bar(df_ciudades['city'], df_ciudades['population_total'], color= ['blue', 'grey', 'red','olive', 'brown'], 
        alpha= 0.5)) # creo el ax y almaceno las barras en una variable para mas adelante
ax1.set_title('Ciudades mas pobladas de California', fontsize= 10)
ax1.set_ylabel('Habitantes (miles)')
ax1.set_xlabel('Ciudades')
ax1.yaxis.set_major_formatter(formatter)
ax1.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:.0f}'.format(x/1000)));

for bar in bars: # coloco las cifras sobre las barras como un texto
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width() / 2, height + 0.5, str(height), ha='center', va='bottom')
```

ax2 arr-der

```python
font_dict = {
    'size': 18,
    'family': "Bradley Hand itc",
}
ax2 = ax[0, 1]
ax2.hist(df_limpio['area_land_sq_mi'], bins= 50, alpha= 0.7, color= 'steelblue')
ax2.set_ylabel('Numero de ciudades', fontdict= font_dict)
ax2.set_xlabel('Superficie $km^2$', fontdict= font_dict)
ax2.legend(['Hist Area'], fontsize= 9);
```

ax3 aba-izq

```python
# preparo las flechas

#NO ME COGE EL ARROWPROPS
arrowprops = {
    'arrowstyle': '->',
    'connectionstyle': 'arc3'
}

x_arrow_1 = df_limpio['area_water_km2'].max() - 73
y_arrow_1 = df_limpio.set_index('population_total')['area_water_km2'].idxmax() - 3300000

x_arrow_2 = df_limpio.set_index('area_water_km2')['population_total'].idxmax() - 73
y_arrow_2 = df_limpio['population_total'].max() - 3300000
```

```python
print(x_arrow_2) # comprobaciones de las arrows, anotaciones personales
print(y_arrow_2)
```

```python
formatter = ticker.ScalarFormatter(useOffset=False, useMathText=True)
formatter.set_scientific(False)  # Desactivar la notación científica

ax3 = ax[1, 0]
ax3.scatter(df_limpio['area_water_km2'], df_limpio['population_total'], 
            alpha= 0.3, color= 'green')
ax3.set_title('Poblacion por kilometro cuadrado de agua', fontsize= 10)
ax3.set_xlabel('Area de agua en la ciudad (km$^2$)')
ax3.set_ylabel('Poblacion (miles de habitantes)')
ax3.text(60, 3300000, 'Outliers', fontsize= 8)
ax3.arrow(73, 3300000, dx= x_arrow_1, dy= y_arrow_1)
ax3.arrow(73, 3300000, dx= x_arrow_2, dy= y_arrow_2)
ax3.yaxis.set_major_formatter(formatter)
ax3.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:.0f}'.format(x/1000)));
```

ax4 aba-der

```python
boxprops = {
    'color': 'black',
    'facecolor': 'red'
}
medianprops = {
    'color': 'black'
}
```

```python
ax4 = ax[1, 1]
ax4.boxplot(df_limpio['elevation_m'], whis= 1.5, vert= False, boxprops= boxprops, 
            medianprops= medianprops, patch_artist= True)
ax4.set_title('Diagrama de caja de la altitud de las ciudades de california', fontsize= 8)
ax4.set_xlabel('Altitud (m)', fontsize= 8);
```
