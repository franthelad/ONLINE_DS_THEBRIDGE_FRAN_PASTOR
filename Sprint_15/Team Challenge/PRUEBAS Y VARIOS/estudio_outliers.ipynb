```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
train_set = pd.read_csv('./data/train_limpio.csv').drop(['Price_euros'], axis= 1)
```

```python
train_set
```

```python
for i in train_set.columns:
    sns.boxplot(train_set[i])
    plt.show()
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='median')

train_set_imputed = imputer.fit_transform(train_set)
```

```python
train_set_imputed = pd.DataFrame(train_set_imputed, columns= train_set.columns, index= train_set.index)
```

```python
for i in train_set_imputed.columns:
    sns.boxplot(train_set_imputed[i])
    plt.show()
```
